Exploracao inicial usando o SQLite local carregado a partir dos CSVs canonicos.

```python
import sqlite3
from pathlib import Path

conn = sqlite3.connect(Path('../db/banco.db'))
conn.execute('PRAGMA foreign_keys = ON')
```

```python
queries = {
    'clientes_por_segmento': "SELECT segmento, COUNT(*) AS total FROM clientes GROUP BY segmento ORDER BY total DESC",
    'saldo_medio_por_tipo': "SELECT tipo_conta, AVG(saldo_atual) AS saldo_medio FROM contas GROUP BY tipo_conta ORDER BY saldo_medio DESC",
    'volume_mensal': "SELECT substr(data_transacao, 1, 7) AS ano_mes, SUM(valor) AS volume_total FROM transacoes GROUP BY substr(data_transacao, 1, 7) ORDER BY ano_mes",
    'tickets_por_categoria': "SELECT categoria_ticket, COUNT(*) AS total FROM tickets_atendimento GROUP BY categoria_ticket ORDER BY total DESC"
}

for name, query in queries.items():
    print(f'--- {name} ---')
    for row in conn.execute(query).fetchmany(10):
        print(row)
    print()
```
